```python
sns.barplot(x='day', y='total_bill', data=data_tips, hue='time', palette='dark:#ffcc99')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(y='day', x='total_bill', data=data_tips, hue='time', palette='dark:#ffcc99',errorbar=None)

# Add labels on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type='edge')  # '%.2f' displays numbers with 2 decimals

# Move the legend
plt.legend(loc='lower right')  # You can use 'best', 'lower right', 'upper right', etc.

plt.show()
```

### Explanation:
1. for container in ax.containers: 
   This loop iterates over the containers (groups of bars) in the Axes object (`ax`). In Matplotlib, each set of bars in a bar chart is stored in a container.
2. ax.bar_label(container, fmt="%.2f", label_type='edge') 
   - ax.bar_label(container): This method is used to attach labels to the bars in the specified container.
   - fmt="%.2f": The `fmt` parameter defines the format of the labels. Here, `"%.2f"` specifies that the labels should be displayed as floating-point numbers with 2 decimal places.
   - label_type='edge': This parameter determines where the labels will be positioned relative to the bars. Using `'edge'` places the labels at the edge of the bar. Other options include `'center'`.

## 3.Box plot

![Boxplot](Boxplot.webp)


```python
plt.figure(figsize=(14,9))
sns.set_style('darkgrid')

# A box plot allows you to compare different variables
# The box shows the quartiles of the data. The bar in the middle is the median and
# the box extends 1 standard deviation from the median
# The whiskers extend to all the other data aside from the points that are considered
# to be outliers
# Hue can add another category being sex
# We see men spend way more on Friday versus less than women on Saturday
sns.boxplot(x='day',y='total_bill',data=data_tips, hue='sex')

# Moves legend to the best position
plt.legend(loc=0)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data_tips, hue='day', palette='dark:#ffcc99', legend=False)
plt.title('Total Bill by Day')
plt.show()
```

### Strip Plot

```python
plt.figure(figsize=(8,5))

# The strip plot draws a scatter plot representing all data points where one
# variable is categorical. It is often used to show all observations with 
# a box plot that represents the average distribution
# Jitter spreads data points out so that they aren't stacked on top of each other
# Hue breaks data into men and women
# Dodge separates the men and women data
sns.stripplot(x='day',y='total_bill',data=data_tips, jitter=True, 
              hue='sex', dodge=True)
plt.show()
```

```python
#plt.figure(figsize=(10, 6))
sns.stripplot(x='day', y='total_bill', data=data_tips, hue='sex', palette='viridis',legend=True)
plt.title('Total Bill by Day')
plt.legend(loc='upper left')
plt.show()
```

### Palettes

```python
plt.figure(figsize=(8,6))

sns.set_style('dark')

sns.set_context('talk')

# You can use Matplotlibs color maps for color styling
# https://matplotlib.org/3.3.1/tutorials/colors/colormaps.html
sns.stripplot(x='day',y='total_bill',data=data_tips, hue='sex', 
             palette='seismic')

# Add the optional legend with a location number (best: 0, 
# upper right: 1, upper left: 2, lower left: 3, lower right: 4,
# https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.legend.html)
# or supply a tuple of x & y from lower left
plt.legend(loc=0)
plt.show()
```

### Swarn plot

```python
# A swarm plot is like a strip plot, but points are adjusted so they don't overlap
# It looks like a combination of the violin and strip plots
# sns.swarmplot(x='day',y='total_bill',data=data_tips)

# You can stack a violin plot with a swarm
sns.violinplot(x='day',y='total_bill',data=data_tips)
sns.swarmplot(x='day',y='total_bill',data=data_tips, color='white')
plt.show()
```

```python
# Create a categorical plot
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=data_tips,palette = 'rocket')
plt.title('Total Bill by Day and Sex')
plt.show()
```

### volin plot

```python
# Violin Plot is a combination of the boxplot and KDE
# While a box plot corresponds to data points, the violin plot uses the KDE estimation
# of the data points
# Split allows you to compare how the categories compare to each other
sns.violinplot(x='day',y='total_bill',data=data_tips, hue='sex',split=True)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=data_tips,hue='sex',palette='mako')
plt.title('Total Bill Distribution by Day')
plt.show()
```

### Joint Plot

```python
# Jointplot compares 2 distributions and plots a scatter plot by default
# As we can see as people tend to speed they also tend to drink & drive
# With kind you can create a regression line with kind='reg'
# You can create a 2D KDE with kind='kde'
# Kernal Density Estimation estimates the distribution of data
# You can create a hexagon distribution with kind='hex'
sns.jointplot(x='speeding', y='alcohol', data=data_crash, kind='reg')
plt.show()
```

### Matrix Plots

### Heatmaps

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

# We can create a matrix with an index of month, columns representing years
# and the number of passengers for each
# We see that flights have increased over time and that most people travel in
# July and August
flights = sns.load_dataset("flights")
flights = flights.pivot_table(index='month', columns='year', values='passengers')
# You can separate data with lines
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1)
plt.show()
```

```python
# Pivot the data and round off the values
data_pivot = data_flights.pivot(index='month', columns='year', values='passengers').round()

# Create a heatmap
sns.heatmap(data_pivot, cmap='viridis', annot=True, fmt='g')  # 'fmt' ensures annotations display as rounded numbers
plt.title('Heatmap of Passengers')
plt.show()
```
